```python
height = 1.6  # meters
weight = 40    # kg
bmi = weight / (height ** 2)

if bmi < 18.5:
    print(f"BMI: {bmi:.1f} - Underweight")
    print("Consider consulting a nutritionist")
elif bmi < 25:
    print(f"BMI: {bmi:.1f} - Normal weight")
    print("Great! Maintain your healthy lifestyle")
elif bmi < 30:
    print(f"BMI: {bmi:.1f} - Overweight")
    print("Consider a balanced diet and exercise")
elif bmi < 35:
    print(f"BMI: {bmi:.1f} - Obese Class I")
    print("Consult a healthcare professional")
elif bmi < 40:
    print(f"BMI: {bmi:.1f} - Obese Class II")
    print("Medical intervention recommended")
else:
    print(f"BMI: {bmi:.1f} - Obese Class III")
    print("Urgent medical consultation required")
```
